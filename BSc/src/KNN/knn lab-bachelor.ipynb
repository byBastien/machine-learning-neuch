## Generate some data points

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import statistics


# Generate normally distributed data with mean 0 and standard deviation 1, 100 samples, 3 classes and 2 features.
n_samples=100
class_prob = [0.33, 0.33, 0.34]
x = np.random.normal(loc = 0.0, scale= 1.0, size = (n_samples, 2))
y = np.random.choice(3, p=class_prob, size=n_samples)


# Convert the feature array `x` to a DataFrame with columns 'feature1' and 'feature2'
x = pd.DataFrame(data=x, columns=['feature1', 'feature2'])

# Convert the target array `y` to a DataFrame with column 'target'
y = pd.DataFrame(data=y, columns=['target'])

# Combine the feature DataFrame and target DataFrame into a single DataFrame
data = pd.concat([x, y], axis=1)

# Display the first few rows of the combined DataFrame
data.head()
```

## Visualize the  data

```python
palette = { 0:'green', 1:'blue', 2:'red'}

colors = ['r','darkgreen','b','c','m','y']

# Create a figure with specified size
fig = plt.figure(figsize=(15, 8))

# Create the scatter plot
sns.scatterplot(data=data, x='feature1', y='feature2', hue='target', 
                alpha=0.9, s=250, palette=palette)


# Add legend with proper labels
ax = fig.get_axes()[0]
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, title='Class', title_fontsize='13', loc='upper right')

# Display the plot
plt.show()
```

## Randomly create a new data point or test data

```python
# Generate a random value for x1 within the range of feature1
x1 = random.uniform(data.feature1.min(), data.feature1.max())

# Generate a random value for x2 within the range of feature2
x2 = random.uniform(data.feature2.min(), data.feature2.max())

# Create the test point
test_data = np.array([x1, x2])

# Create the scatter plot
fig = plt.figure(figsize=(15, 8))
sns.scatterplot(data=data, x='feature1', y='feature2', hue='target', 
                alpha=0.9, s=250, palette=palette)

plt.scatter(x=x1, y=x2, s=250, marker='*', color='k', label='test_data')

# Add legend with proper labels
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, labels, title='Class', title_fontsize='13', loc='upper right')

# Set titles for x and y axes
plt.xlabel('Feature 1', fontsize=14)
plt.ylabel('Feature 2', fontsize=14)

# Display the plot
plt.show()
```
